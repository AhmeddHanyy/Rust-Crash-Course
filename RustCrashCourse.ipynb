## Variables in Rust

### Basic Variable Declaration

Variables in Rust are declared using the `let` keyword. By default, variables are **immutable** (cannot be changed after initialization). This is a key design principle in Rust that prevents bugs.

```rust
// Basic variable declaration
let x = 5;
println!("x = {}", x);

// x = 10;  // ❌ This would cause a compilation error!
```

### Mutable Variables

To make a variable mutable, use the `mut` keyword. This explicitly declares your intent to change the variable's value.

```rust
// Mutable variable declaration
let mut y = 5;
println!("y = {}", y);

y = 10;  // ✅ This works because y is mutable
println!("y = {}", y);

y = 15;
println!("y = {}", y);
```

### Type Annotations

Rust is a statically typed language, which means that it must know the types of all variables at compile time. It can infer types, but you can explicitly annotate them using the syntax: `let variable: type = value;`

```rust
// Type annotations
let name: &str = "Alice"; //Will talk about strings later
let age: u32 = 25;
let height: f64 = 5.6;
let is_student: bool = true;

println!("Name: {}", name);
println!("Age: {}", age);
println!("Height: {}", height);
println!("Is student: {}", is_student);

/*
As shown we can declare a variable with or without stating its type. Rust can infer the type based on the value assigned to it.
However, sometimes it's necessary to provide a type annotation, especially when the type cannot be inferred or when you want to specify a different type than what Rust would infer by default.
*/
```

### Common Data Types

| Type | Description | Example |
|------|-------------|----------|
| `i32` | Signed 32-bit integer | `let x: i32 = -42;` |
| `u32` | Unsigned 32-bit integer | `let x: u32 = 42;` |
| `f64` | 64-bit floating-point | `let x: f64 = 3.14;` |
| `bool` | Boolean (true/false) | `let x: bool = true;` |
| `&str` | String slice (immutable) | `let x: &str = "hello";` |
| `String` | Owned string | `let x: String = String::from("hello");` |
| `char` | Single Unicode character | `let x: char = 'A';` |

```rust
// Examples of different types
let x: i32 = -10;
let y: u32 = 10;
let z: f64 = 10.5;
let flag: bool = false;
let letter: char = 'Z';

println!("i32: {}", x);
println!("u32: {}", y);
println!("f64: {}", z);
println!("bool: {}", flag);
println!("char: {}", letter);
```

### Variable Shadowing

Shadowing allows you to redeclare a variable with the same name. The new variable "shadows" the previous one. This is different from mutability - you can change the type when shadowing!

```rust
// Variable shadowing
let x = 5;
println!("x = {}", x);

let x = x + 1;  // Shadowing: x is now 6
println!("x = {}", x);

let x = "hello";  // Shadowing: x changes to a string [We changed the type here]
println!("x = {}", x);
```

### Constants

Constants are values that cannot be changed. They are declared using the `const` keyword and **require explicit type annotations**. By convention, constants are written in ALL_CAPS.

```rust
// Constants - must have explicit type annotation
const MAX_POINTS: u32 = 100;
const PI: f64 = 3.14159;
const GRAVITY: f64 = 9.8;

println!("Max points: {}", MAX_POINTS);
println!("Pi: {}", PI);
println!("Gravity: {}", GRAVITY);

// Constants are evaluated at compile time and can be used in const contexts
```

### Key Differences: Mutability vs Shadowing

| Feature | Mutability | Shadowing |
|---------|-----------|----------|
| How it works | Change value with `mut` | Redeclare with `let` |
| Can change type? | No | Yes ✓ |
| Syntax | `let mut x = 5; x = 10;` | `let x = 5; let x = "hello";` |
| Scope | Remains same | Can start new scope |
| Use case | Modify value in place | Transform value, change type |

```rust
// Example: Mutability vs Shadowing

// Using mutability - same type
let mut score = "100";
// score = 105;  // ❌ Error: type mismatch, remains string

// Using shadowing - can change type  
let mut value = "100";
let value = value.parse::<i32>().unwrap();  // ✓ Now it's an integer!
println!("Value as int: {}", value);
```

Important Note: All the above types (ignoring strings) are stack allocated as it previously knows the exact size. We'll talk about heap allocations later.

## Functions in Rust

Functions are one of the most important building blocks in Rust. They help organize code, promote reusability, and make programs more maintainable. Let's explore how functions work in Rust.

### Basic Function Declaration

Functions are declared using the `fn` keyword. A simple function takes no parameters and returns nothing (the `()` type, also called "unit").

```rust
// Basic function declaration
fn greet() {
    println!("Hello, Rust!");
}

// Call the function
greet();
greet();
```

### Functions with Parameters

Functions can accept parameters. Each parameter requires an EXPLICIT type annotation. Parameters are separated by commas.

```rust
// Function with parameters
fn add(x: i32, y: i32) {
    let result = x + y;
    println!("{} + {} = {}", x, y, result);
}

add(5, 3);
add(10, 20);
```

### Return Values

Functions can return values using the `->` syntax to specify the return type. You return a value using either:
- An explicit `return` statement (with semicolon)
- The last expression **without a semicolon** (this is the idiomatic way)

```rust
// Function with return value - explicit return
fn multiply_explicit(x: i32, y: i32) -> i32 {
    return x * y;  // Note the semicolon
}

// Function with return value - implicit (idiomatic)
fn multiply(x: i32, y: i32) -> i32 {
    x * y  // No semicolon! This is the return value
}

println!("5 * 3 = {}", multiply(5, 3));
println!("10 * 2 = {}", multiply(10, 2));
```

### Statements vs Expressions

This is a key Rust concept!

- **Statements** end with a semicolon and return nothing (the unit type `()`)
- **Expressions** do NOT end with a semicolon and return their value

In the function above, `x * y` is an expression that returns a value, but `return x * y;` is a statement.

### Key Differences: Statement vs Expression

| Feature | Statement | Expression |
|---------|-----------|-----------|
| Ends with | Semicolon `;` | No semicolon |
| Returns value? | No (returns `()`) | Yes |
| Example | `let x = 5;` | `5 + 6` or `{ let x = 3; x + 1 }` |
| In function | `return value;` | `value` |
| Usage | Performs action | Computes and returns value |

```rust
// Expressions in if statements
fn check_positive(x: i32) -> &'static str {
    if x > 0 {
        "positive"  // expression - no semicolon
    } else if x < 0 {
        "negative"  // expression - no semicolon
    } else {
        "zero"      // expression - no semicolon
    }
}

println!("5 is {}", check_positive(5));
println!("-3 is {}", check_positive(-3));
println!("0 is {}", check_positive(0));
```

### Multiple Return Values (Tuples)

Functions can return multiple values using tuples. But first let's see what a tuple is.

A tuple is a general way of grouping together a number of values with a variety of types into one compound type. Tuples have a fixed length: Once declared, they cannot grow or shrink in size.

```rust
let tup = (500,'a',1.5);
println!("{}",tup.1);
println!("{first}, {second}, {third}", 
    first = tup.0, 
    second = tup.1, 
    third = tup.2
);
println!("{:?}", tup);
let (x, y, z) = tup;
let empty : () = (); //empty tuple - when functions return nothing they return empty tuple
```

Let's return back to functions returning multiple values

```rust
// Function returning a tuple
fn divide_and_remainder(dividend: i32, divisor: i32) -> (i32, i32) {
    (dividend / divisor, dividend % divisor)    
}

let (quotient, remainder) = divide_and_remainder(17, 5);
println!("17 / 5 = {} with remainder {}", quotient, remainder);

let result = divide_and_remainder(20, 3);
println!("20 / 3 = {} remainder {}", result.0, result.1);
```

### Function Example: Complete Program

Let's create a useful function that calculates the area and perimeter of a rectangle.

```rust
// Calculate area and perimeter of a rectangle
fn rectangle_stats(width: f64, height: f64) -> (f64, f64) {
    let area = width * height;
    let perimeter = 2.0 * (width + height);
    (area, perimeter)
}

// Using the function
let (area, perimeter) = rectangle_stats(5.0, 3.0);
println!("Rectangle 5x3:");
println!("  Area: {}", area);
println!("  Perimeter: {}", perimeter);

let stats = rectangle_stats(10.0, 8.0);
println!("\nRectangle 10x8:");
println!("  Area: {}", stats.0);
println!("  Perimeter: {}", stats.1);
```

## Control Flow

Control flow is how we direct the path a program takes. Rust provides several ways to control program flow: if/else conditions, match expressions, and loops. Let's explore each one.

### if/else Statements

The `if` keyword allows you to conditionally execute code based on a boolean condition. Unlike other languages, Rust **requires** the condition to be a boolean—it won't automatically convert other types.

```rust
// Simple if statement
let number = 6;

if number < 5 {
    println!("Number is less than 5");
} else if number == 5 {
    println!("Number equals 5");
} else {
    println!("Number is greater than 5");
}

// if as an expression (the idiomatic way)
let age = 25;
let status = if age >= 18 { "adult" } else { "minor" };
println!("Age {} is: {}", age, status);

/*
if 5 {
    println!("This will not compile because 5 is not a boolean expression");
}
*/
```

### match Statements (Rust's Switch)

`match` is Rust's powerful pattern matching construct and is much more flexible than switch statements in other languages. It **must be exhaustive** (all possible values must be handled). It's often used instead of if/else for cleaner code.

```rust
// Basic match statement
let day = 3;

match day {
    1 => println!("Sunday"),
    2 => println!("Monday"),
    3 => println!("Tuesday"),
    4 => println!("Wednesday"),
    5 => println!("Thursday"),
    6 => println!("Friday"),
    7 => println!("Saturday"),
    _ => println!("Invalid day"), // catch-all case
}

// match as an expression
let color = "red";
let hex_value = match color {
    "red" => "#FF0000",
    "green" => "#00FF00",
    "blue" => "#0000FF",
    _ => "#000000", // black as default
};
println!("Color {} is {}", color, hex_value);

let x = 10;
match x {
    10 => {
        println!("x is ten");
        // We can have multiple statements here
        let y = x * 2;
        println!("y is {}", y);
    }
    _ => println!("x is not ten"),
}
```

### if let: Concise Pattern Matching

`if let` is a shorthand for `match` when you only care about one specific pattern and want to ignore the rest. It's much cleaner and more readable when you don't need to handle all cases.

Syntax: `if let pattern = expression { /* handle match */ } else { /* handle no match */ }`

This is especially useful when working with `Option` and `Result` types (which we'll cover later).


```rust
// if let vs match comparison

// With match (verbose when you only care about one case)
let value = Some(5);
match value {
    Some(x) => println!("Got a value: {}", x),
    None => println!("Got nothing"),
}

// Same with if let (much cleaner!)
if let Some(x) = value {
    println!("Got a value: {}", x);
} else {
    println!("Got nothing");
}

// More practical example
let config_value = Some(42);

if let Some(val) = config_value {
    println!("Configuration value is: {}", val);
} else {
    println!("No configuration value set");
}

// You can also omit the else clause if you don't need it
let number = Some(10);
if let Some(n) = number {
    println!("Number: {}", n);
}
// No else needed - if the pattern doesn't match, just continue

// Useful in loops too!
let values = vec![Some(1), None, Some(3), Some(4), None];
for val in values {
    if let Some(n) = val {
        println!("Processing: {}", n);
    }
    // If None, we just skip the print
}
```

### let else: Early Return Pattern

`let else` is a modern Rust pattern that lets you extract a value from a pattern, and if the pattern doesn't match, execute an else block (usually to return or bail out early).

Syntax: `let pattern = expression else { /* early exit */ };`

This is useful for **error handling and validation** - you extract the success case, and immediately handle the failure case without deep nesting.


```rust
// let else example

fn process_number(value: Option<i32>) {
    // Using let else for early return
    let Some(num) = value else {
        println!("Error: No value provided");
        return;
    };

    // If we reach here, num is guaranteed to be i32
    println!("Processing number: {}", num);
    println!("Double: {}", num * 2);
}

process_number(Some(5));
process_number(None);

println!();

// Practical example: Parsing and validation
fn validate_age(age_str: &str) -> String {
    // Try to parse into i32
    let Ok(age) = age_str.parse::<i32>() else {
        return String::from("Invalid input - not a number");
    };

    // Try to check if age is valid
    if age < 0 || age > 150 {
        return String::from("Invalid age - must be 0-150");
    }

    format!("Valid age: {}", age)
}

println!("{}", validate_age("25"));
println!("{}", validate_age("invalid"));
println!("{}", validate_age("200"));

println!();

// In multiple assignments
fn extract_values() {
    let Some(x) = Some(10) else {
        println!("x not found!");
        return;
    };

    let Some(y) = Some(20) else {
        println!("y not found!");
        return;
    };

    println!("x = {}, y = {}, sum = {}", x, y, x + y);
}

extract_values();
```

### Comparison: match vs if let vs let else

| Construct | Best For | Example |
|-----------|----------|---------|
| `match` | Handling multiple patterns exhaustively | Multiple arms with different behaviors |
| `if let` | Handling ONE pattern, optionally with else | Checking `Some(x)` or a specific variant |
| `let else` | Extracting a value with early exit on failure | Validation, error handling, guards |

**Quick Guide:**
- Use **`match`** when you have 2+ patterns to handle
- Use **`if let`** when you want one specific pattern and optionally an else block
- Use **`let else`** when you want to extract a value and bail out early if it fails

All three deal with **pattern matching**, but they're optimized for different scenarios!


### Loop: Infinite Loop

The `loop` keyword creates an infinite loop that continues until explicitly broken with the `break` keyword(will come to it right after).

```rust
// Infinite loop with break
let mut counter = 0;

loop {
    counter += 1;
    println!("Counter: {}", counter);
    
    if counter == 5 {
        break;  // Exit the loop
    }
}

// loop as an expression - returning a value
let mut num = 0;
let result = loop {
    num += 1;
    if num == 10 {
        break num * 2;  // Break with a value to return
    }
};
println!("Loop returned: {}", result);
```

### while Loops

A `while` loop repeats as long as a condition is true.

```rust
// Basic while loop
let mut count = 1;

while count <= 5 {
    println!("Count: {}", count);
    count += 1;
}

// While with complex conditions
let mut x = 10;
while x > 0 && x <= 10 {
    println!("x = {}", x);
    x -= 2;
}
```

### for Loops

A `for` loop is the most common loop in Rust. It iterates over elements in a collection (arrays, ranges, etc.).

```rust
// For loop with a range
for i in 1..=5 {  // 1..=5 means 1 to 5 inclusive
    println!("i = {}", i);
}
// Note that i doesn't need to be explicitly declared.
// For loop with increment by 2
for i in (1..=10).step_by(2) {
    println!("i = {}", i);
}

for k in 1..=5 {
    println!("k = {}", k);
}
//println!("k = {}", k); this will error out as k scope is finished
// For loop over an array -> array ??
let arr: [i32; 5] = [10, 20, 30, 40, 50]; // arrays are just fixed-size collections of elements of the same type very similar to tuples but all elements must be of the same type and they have a fixed size known at compile time

let numbers = [10, 20, 30, 40];
for num in numbers { //note num is declared implicitly as i32 here as numbers is an array of i32
    println!("Number: {}", num);
}

// For loop with index (enumerate)
let fruits = ["apple", "banana", "orange"];
for (index, fruit) in fruits.iter().enumerate() {
    println!("Index {}: {}", index, fruit);
}
```

### break and continue

- `break` exits the loop immediately
- `continue` skips to the next iteration

```rust
// break example
println!("Break example:");
for i in 1..10 {
    if i == 5 {
        println!("  Reached 5, breaking out");
        break;
    }
    println!("  i = {}", i);
}

// continue example
println!("\nContinue example:");
for i in 1..=5 {
    if i == 3 {
        println!("  Skipping 3");
        continue;
    }
    println!("  i = {}", i);
}
```

## Ownership

Ownership is **the most important concept in Rust**. It's what makes Rust memory-safe without a garbage collector. Understanding ownership is crucial to becoming proficient with Rust.

The ownership system governs how memory is allocated and deallocated. Every value in Rust has an owner, and when the owner goes out of scope, the value is dropped (deallocated).

### Stack vs Heap

Before diving into ownership, let's understand where data is stored:

- **Stack**: Fixed-size, known at compile time, fast access, automatically cleaned up. Examples: integers, booleans, floats, chars
- **Heap**: Variable-size, determined at runtime, slower access, requires manual (or ownership-based) cleanup. Examples: String, Vec, etc.


### Stack-Allocated Types and Copy Trait

Types like integers, floats, booleans, and chars are stored on the stack. These types have a special property: they implement the **Copy trait** (we'll talk about traits later).

The Copy trait means that Rust **copies** the value from an variable to another without affecting the source in any way. This is why you can use the same integer variable multiple times without issues.


```rust
// Stack-allocated types (Copy trait)
// You can use these values multiple times without any issues

let x = 5;
println!("x = {}", x);  // x is still valid here
println!("x = {}", x);  // and here
println!("x = {}", x);  // and here too!

let y = x;  // y gets a COPY of x's value
println!("x = {}, y = {}", x, y);  // Both x and y are still valid!

// This works because integers (and other stack-allocated types [not the user-defined types])
// implement the Copy trait. The value is duplicated, not moved.

let a = 10;
let b = a;  // b gets a copy of a
let c = a;  // c also gets a copy of a
println!("a = {}, b = {}, c = {}", a, b, c);  // All are still valid!
```

### Heap-Allocated Types and Move Semantics

Unlike integers, types like `String` are stored on the **heap**. They contain a pointer to the data on the heap, plus metadata (length and capacity).

When you assign a heap-allocated value to another variable, Rust **moves** the ownership instead of copying. The original variable becomes invalid after the move.


```rust
// Heap-allocated types (Move semantics)
// When you assign a String to another variable, ownership MOVES

let s1 = String::from("Hello"); //'String' is a struct, we'll mention structs later. from() is a method implemented for this struct
println!("s1 = {}", s1);

let s2 = s1;  // s1's ownership is MOVED to s2
// s1 is now invalid!

println!("s2 = {}", s2);  // ✓ This works

// println!("s1 = {}", s1);  // ❌ ERROR! s1 is no longer valid
// error[E0382]: borrow of moved value: `s1`

// The String's data still exists on the heap, but only s2 owns it now.
// When s2 goes out of scope, the String's memory will be freed.
```

```rust
// More practical examples of moves

let name1 = String::from("Alice");
let name2 = name1;  // Move: name1 is now invalid

// We can reassign name1 with a new String
let name1 = String::from("Bob");

println!("name1 = {}", name1);  // ✓ Bob is valid
println!("name2 = {}", name2);  // ✓ Alice is owned by name2

// When name2 goes out of scope at the end, Alice's memory is freed
// When name1 goes out of scope, Bob's memory is freed

/* 
When asked why move happens, we have two reasons: first the type used which is String does not implement the Copy trait,
and second to prevent double free errors.
If both name1 and name2 were valid and owned the same String data, when they go out of scope they would both try to free
the same memory, leading to a double free error.
By moving ownership, Rust ensures that only one variable is responsible for freeing the memory, thus preventing this issue.
*/
```

### Ownership with Functions

Ownership also works with function parameters. When you pass a value to a function:
- **Stack-allocated types (Copy)**: The value is **copied**, and both the caller and function have valid independent values
- **Heap-allocated types**: Ownership is **moved**, and the original variable becomes invalid after the function call

This creates challenges! Let's see why...


```rust
// Copy trait works seamlessly with functions
// Copy types are passed by value and duplicated

fn print_number(num: i32) {
    println!("Inside function: {}", num);
    // num will not be valid outside this function, but since it's a Copy type, the caller's variable remains valid
}

let x = 42;
print_number(x);      // x is COPIED, not moved
println!("x = {}", x);  // ✓ x is still valid!
print_number(x);      // ✓ We can use x again!

// This works great because integers implement Copy.
// The value is duplicated, so x remains valid in the caller.
```

```rust
// Now let's try with String - this is where problems arise!

fn print_string(s: String) {
    println!("Inside function: {}", s);
    // When print_string ends, s goes out of scope and is dropped (freed). Btw freeing happens internally in Rust by calling drop() on the String, which deallocates the heap memory. This is why we can't use s after this function ends.
}

let message = String::from("Hello");
print_string(message);  // message's ownership is MOVED into the function

// println!("message = {}", message);  // ❌ ERROR! message is no longer valid
// error[E0382]: borrow of moved value: `message`

// The String is freed when print_string ends, and we can't use it anymore.
// This is very inconvenient if we need to use the String multiple times!
```

```rust
// One solution: Return ownership back from the function!
// But this gets messy quickly...

fn print_and_return(s: String) -> String {
    println!("Inside function: {}", s);
    s  // Return ownership back
}

let msg = String::from("Important");
let msg = print_and_return(msg);  // Must reassign each time
println!("After function: {}", msg);

// This works, but it's awkward and verbose!
// What if we call multiple functions?

fn append_exclamation(mut s: String) -> String {
    s.push_str("!");
    s
}

fn make_uppercase(mut s: String) -> String {
    let s_upper = s.to_uppercase();
    s_upper
}

let text = String::from("hello");
let text = append_exclamation(text);        // Have to reassign
let text = make_uppercase(text);            // Have to reassign again
println!("Result: {}", text);

// This is tedious and error-prone!
// You have to constantly move ownership back and forth.
// This is exactly the problem that references solve!
```

### The Problem: Life Without References

As you can see, working with ownership for heap-allocated types (like String) gets tedious quickly:

1. **Ownership moves into functions** - the caller loses access to the value
2. **You must return ownership back** - adding explicit return statements and reassignments
3. **Code becomes verbose** - every function call requires tracking ownership
4. **Hard to write flexible functions** - functions can't both use AND return values cleanly

**This is where references come in!** References allow you to let a function *borrow* a value without taking ownership. They're the key to writing clean, practical Rust code. We'll cover them next!

### Summary: Ownership Rules

1. Each value in Rust has an owner
2. There can only be one owner at a time
3. When the owner goes out of scope, the value is dropped
4. **Copy types** (i32, bool, f64, char) are copied when assigned/passed to functions
5. **Non-Copy types** (String, Vec) are moved when assigned/passed to functions
6. Moving ownership gives us automatic memory management without garbage collection


## Borrowing and References

References allow a function to use a value without taking ownership of it. This is called **borrowing**. When you borrow a value, you get a reference to it (denoted by `&`). The original owner still owns the value, and the borrower can't accidentally deallocate it!

This solves the ownership transfer problem we saw earlier.

### Immutable References: `&T`

An immutable reference allows you to **read** a value but not modify it. You create an immutable reference using the `&` operator. Multiple immutable references can exist at the same time.


```rust
// Immutable references - read-only borrowing

fn print_string(s: &String) {
    println!("Inside function: {}", s);
    // We can read s, but not modify it
    // s is a reference, so it's not dropped here
}

let message = String::from("Hello");
print_string(&message);      // Borrow immutably with &message
println!("message = {}", message);  // ✓ message is still valid!
print_string(&message);      // ✓ We can borrow again!

// The String's ownership never moved to the function.
// The function just borrowed it temporarily.
```

```rust
// Multiple immutable references at the same time

fn get_length(s: &String) -> usize {
    s.len()
}

fn print_ref(s: &String) {
    println!("String: {}", s);
}

let text = String::from("Rust");

// We can create multiple immutable references
let len = get_length(&text);
print_ref(&text);
print_ref(&text);

println!("Text: '{}', Length: {}", text, len);

// All these references are borrowing the same value.
// text still owns the String and remains valid!
```

### Mutable References: `&mut T`

A mutable reference allows a function to **read and modify** a value without taking ownership. You create a mutable reference using `&mut`.

**Important Rule**: You can only have **ONE** mutable reference at a time. This prevents data races and ensures safety.


```rust
// Mutable references - read and write borrowing

fn add_exclamation(s: &mut String) {
    s.push_str("!");
}

fn append_period(s: &mut String) {
    s.push('.');
}

let mut greeting = String::from("Hello");
println!("Original: {}", greeting);

add_exclamation(&mut greeting);  // Borrow mutably
println!("After exclamation: {}", greeting);

append_period(&mut greeting);
println!("After period: {}", greeting);

// The String ownership never moved!
// Functions can modify it without taking ownership.
```

```rust
// The One Mutable Reference Rule
// You cannot have multiple mutable references to the same value
{
let mut num = 42;
let mut num1 = 50;

// This is valid - one mutable reference
let ref1 = &mut num;
*ref1 = 100;  // Modify through the reference
println!("num = {}", num);

// ❌ This would NOT compile:
// let ref2 = &mut num;  // ERROR! Can't have a second mutable reference
// *ref1 = 50;           // And we can't use ref1 after ref2 is created
// *ref2 = 75;

// This prevents data races: two functions can't modify the same data simultaneously!

// However, you CAN have multiple IMMUTABLE references:
let ref_a = &num1;
let ref_b = &num1;
let ref_c = &num1;
println!("Via ref_a: {}, ref_b: {}, ref_c: {}", ref_a, ref_b, ref_c);
}
// Multiple readers are safe, but only one writer at a time!
```

### The Power of References: Solving the Original Problem

Remember the tedious ownership-passing code we had to write earlier? Now with references, it becomes clean and simple!


```rust
// Before: Without references (tedious ownership transfer)
fn add_excl_owned(mut s: String) -> String {
    s.push_str("!");
    s
}

fn make_upper_owned(s: String) -> String {
    s.to_uppercase()
}

let text = String::from("hello");
let text = add_excl_owned(text);        // Have to reassign
let text = make_upper_owned(text);      // Have to reassign
println!("Old way: {}", text);

println!();

// After: With references (clean and simple!)
fn add_excl_ref(s: &mut String) {
    s.push_str("!");
}

fn make_upper_ref(s: &String) -> String {
    s.to_uppercase()
}

let mut text2 = String::from("hello");
add_excl_ref(&mut text2);              // No reassignment needed!
let upper = make_upper_ref(&text2);    // Borrow, don't move
println!("New way: {} -> {}", text2, upper);

// Much cleaner! No ownership juggling needed!
```

### The Borrowing Rules

Rust enforces three golden rules for borrowing that ensure memory safety without a garbage collector:

1. **Ownership**: Each value has exactly one owner at a time
2. **Immutable Borrowing**: You can have many immutable references (`&T`) at the same time
3. **Mutable Borrowing**: You can have at most ONE mutable reference (`&mut T`) at a time, and you can't mix immutable and mutable references

The motivation: **Multiple readers are safe, but a writer needs exclusive access.**


```rust
// Demonstrating the borrowing rules
{
let mut data = vec![1, 2, 3]; //will talk about vectors later, for now just know it's a growable array type

// Multiple immutable borrows - ✓ OK
let r1 = &data;
let r2 = &data;
println!("r1: {:?}, r2: {:?}", r1, r2); //the :? is used to print the vector in a debug format

// Now an immutable borrow has ended, we can create a mutable borrow
let r_mut = &mut data;
r_mut.push(4);  // Modify through mutable reference
println!("After mutation: {:?}", r_mut);

// ❌ This would ERROR after creating a mutable borrow:
// let r3 = &data;  // Can't create immutable borrow while mutable borrow exists!

// But this is OK - mutable borrow has ended:
let r3 = &data;
println!("New immutable borrow: {:?}", r3);
}

println!("\n--- Scope Example ---");

let mut x = 5;

{
    let y = &mut x;
    *y = 10;
}  // y's scope ends here

// Now we can borrow again
{
let z = &x;
println!("x = {} (through z)", z);
}
```

### Practical Example: String Manipulation with References

Let's build a practical example combining everything we've learned about references.


```rust
// Helper functions that use references

fn append_string(target: &mut String, addition: &String) {
    target.push_str(addition);  // target is mutable, addition is immutable
}

fn get_length(s: &String) -> usize {
    s.len()
}

fn print_stat(s: &String, label: &str) {
    println!("{}: '{}' (length: {})", label, s, get_length(s));
}

// Using all together
let mut message = String::from("Rust");
append_string(&mut message, &String::from(" is awesome"));
print_stat(&message, "After append");

let word = String::from("!!!");
append_string(&mut message, &word);
print_stat(&message, "After adding word");

// Notice: message can be safely borrowed by multiple functions
// Each function borrows what it needs (mutable or immutable)
// No ownership is transferred, so we keep full control!
```

### Comparison: Different Ways to Pass Values

| Scenario | Syntax | Copy? | Keep Ownership? | Can Modify? | Use Case |
|----------|--------|-------|-----------------|-------------|----------|
| Copy values (i32, etc) | `func(x)` | Yes | Yes | No | Simple data, multiple uses |
| Immutable borrow | `func(&x)` | No | Yes | No | Read-only access |
| Mutable borrow | `func(&mut x)` | No | Yes | Yes | Modify and keep ownership |
| Move ownership | `func(x)` | No | No | Yes | Transfer full control |
| Return ownership | `func() -> T` | No | No | Yes | Transfer ownership back |

### Key Takeaways

1. **References let you borrow without transferring ownership** - The original owner keeps the value
2. **Immutable references (`&T`)** - Multiple threads can read safely
3. **Mutable references (`&mut T`)** - Exclusive access for safe writing
4. **The Borrow Checker** - Rust's compiler enforces these rules at compile time, with no runtime cost
5. **Memory safety without garbage collection** - The rules prevent dangling pointers, data races, and null pointer errors

**References are fundamental to idiomatic Rust.** Understanding them deeply will make you a much better Rust programmer!


### The Dangling Reference Problem

A **dangling reference** occurs when you try to return a reference to data that will be dropped (deallocated) when the function ends. This is a classic memory safety issue in C/C++, but **Rust's borrow checker prevents it at compile time**.

Let's see what happens when we try to create a dangling reference:


```rust
// ❌ DANGLING REFERENCE - This WON'T COMPILE!
// Uncomment to see the error:

/*
fn create_string() -> &String {
    let s = String::from("Hello");
    &s  // ERROR: Trying to return a reference to s
        // But s will be dropped when the function ends!
        // This would leave a dangling pointer
}

let ref_to_danger = create_string();
println!("{}", ref_to_danger);  // ref_to_danger points to freed memory!
*/

// The Rust compiler catches this with an error like:
// error[E0515]: cannot return reference to local variable `s`
//   |
// | fn create_string() -> &String {
// |                       --------
// |                       |
// |                       help: consider using the `'a` lifetime parameter as shown
// |                       return type of function, and adding `'a` as a generic lifetime parameter:
// |                       fn create_string() -> &'a String {
// | 
// | let s = String::from("Hello");
// | ------                         - `s` is borrowed here
// | &s  -- `s` is dropped here while still borrowed
// |
// This is memory safety at compile time!

// The CORRECT solution: Return ownership instead of a reference
fn create_string_correct() -> String {
    let s = String::from("Hello");
    s  // Return the String itself (move ownership)
}

let owned_string = create_string_correct();
println!("Safe: {}", owned_string);

// Now the caller owns the String and can use it as long as needed!
```

## Structs, Methods, and Associated Functions

**Structs** (structures) are custom data types that let you bundle related data together. They're one of the most important building blocks in Rust for organizing code and creating meaningful abstractions.

A struct is similar to a tuple, but structs let you name each piece of data (called **fields**), making your code more readable and maintainable.

### Defining and Creating Structs

To define a struct, you use the `struct` keyword followed by the struct name and its fields with their types.


```rust
// Defining a struct
struct Person {
    name: String,
    age: u32,
    email: String,
}

// Creating an instance of a struct
let person1 = Person {
    name: String::from("Alice"),
    age: 30,
    email: String::from("alice@example.com"),
};

// Accessing struct fields using dot notation
println!("Name: {}", person1.name);
println!("Age: {}", person1.age);
println!("Email: {}", person1.email);

// Creating another instance
let person2 = Person {
    name: String::from("Bob"),
    age: 25,
    email: String::from("bob@example.com"),
};

println!("\n{} is {} years old", person2.name, person2.age);
```

### Mutable Structs

If you need to modify struct fields after creation, you must declare the instance as **mutable** using the `mut` keyword. Note that you can't selectively make only certain fields mutable, all fields become mutable together.


```rust
// Mutable struct
struct Rectangle {
    width: f64,
    height: f64,
}

let mut rect = Rectangle {
    width: 10.0,
    height: 5.0,
};

println!("Original: {} x {}", rect.width, rect.height);

// Modify struct fields
rect.width = 15.0;
rect.height = 8.0;

println!("Modified: {} x {}", rect.width, rect.height);

// You CAN'T do this:
// let rect = Rectangle { ... };  // immutable
// rect.width = 20.0;  // ❌ ERROR! Can't modify immutable struct
```

### Methods: Implementation Blocks

**Methods** are functions associated with a struct. They're defined inside an `impl` (implementation) block. Methods are similar to functions, but:

1. They're defined within an `impl` block
2. Their first parameter is always `self` (representing the instance)
3. They're called using dot notation: `instance.method()`

There are three types of `self`:
- `&self` - borrows the instance immutably (read-only)
- `&mut self` - borrows the instance mutably (can modify)
- `self` - takes ownership of the instance (consumes it)


```rust
// Define a struct
struct Circle {
    radius: f64,
}

// Implement methods for Circle
impl Circle {
    // Method with &self (borrowing immutably)
    // Calculates area
    fn area(&self) -> f64 {
        3.14159 * self.radius * self.radius
    }

    // Method with &self (borrowing immutably)
    // Returns the circumference
    fn circumference(&self) -> f64 {
        2.0 * 3.14159 * self.radius
    }

    // Method with &mut self (borrowing mutably)
    // Modifies the radius
    fn scale(&mut self, factor: f64) {
        self.radius *= factor;
    }

    // Method with self (consumes the instance)
    // This method takes ownership and can't be called again
    fn info(self) -> String {
        format!("Circle with radius: {}", self.radius)
    }
}

// Using methods
let mut circle = Circle { radius: 5.0 };

// Call immutable borrow method
println!("Area: {:.2}", circle.area());
println!("Circumference: {:.2}", circle.circumference());

// Call mutable borrow method
circle.scale(2.0);
println!("After scaling by 2.0: Area = {:.2}", circle.area());

// Call consuming method
let description = circle.info();
println!("{}", description);

// ❌ Can't use circle anymore after info() - it consumed the instance
// println!("{:.2}", circle.area());  // ERROR! circle was moved
```

### Associated Functions

**Associated functions** are functions defined in an `impl` block, but they do **NOT** take `self` as a parameter. They're called using the `::` syntax (e.g., `StructName::function_name()`).

Associated functions are commonly used as **constructors** to create new instances of a struct. You've already seen this with `String::from()`, which is an associated function of the `String` struct!


```rust
// Define a struct
struct Point {
    x: i32,
    y: i32,
}

// Implement methods and associated functions
impl Point {
    // Associated function - constructor (no self parameter)
    fn new(x: i32, y: i32) -> Point {
        Point { x, y }  // Shorthand: x: x becomes x
    }

    // Associated function - another constructor
    fn origin() -> Point {
        Point { x: 0, y: 0 }
    }

    // Regular method (takes &self)
    fn display(&self) {
        println!("Point({}, {})", self.x, self.y);
    }

    // Regular method (takes &self)
    fn distance_from_origin(&self) -> f64 {
        ((self.x * self.x + self.y * self.y) as f64).sqrt()
    }

    // Method (takes &mut self)
    fn translate(&mut self, dx: i32, dy: i32) {
        self.x += dx;
        self.y += dy;
    }
}

// Using associated functions to create instances
let point1 = Point::new(3, 4);
point1.display();
println!("Distance from origin: {:.2}", point1.distance_from_origin());

let mut point2 = Point::origin();
point2.display();

point2.translate(5, 5);
println!("After translating:");
point2.display();

// Similar to how String::from works!
let s = String::from("Hello");
println!("String: {}", s);
```

### Practical Example: A Bank Account Struct

Let's create a comprehensive example combining structs, methods, and associated functions.


```rust
// Bank account struct
struct BankAccount {
    owner: String,
    balance: f64,
}

impl BankAccount {
    // Associated function (constructor)
    fn new(owner: String, initial_balance: f64) -> BankAccount {
        BankAccount {
            owner,
            balance: initial_balance,
        }
    }

    // Method - immutable borrow (read-only)
    fn check_balance(&self) -> f64 {
        self.balance
    }

    // Method - immutable borrow
    fn display_info(&self) {
        println!("Account: {}", self.owner);
        println!("Balance: ${:.2}", self.balance);
    }

    // Method - mutable borrow (modifies state)
    fn deposit(&mut self, amount: f64) {
        if amount > 0.0 {
            self.balance += amount;
            println!("Deposited: ${:.2}", amount);
        }
    }

    // Method - mutable borrow
    fn withdraw(&mut self, amount: f64) -> bool {
        if amount > 0.0 && amount <= self.balance {
            self.balance -= amount;
            println!("Withdrew: ${:.2}", amount);
            true
        } else {
            println!("Insufficient funds or invalid amount");
            false
        }
    }
}

// Using the BankAccount
let mut account = BankAccount::new(String::from("Alice"), 1000.0);
account.display_info();

account.deposit(500.0);
println!("New balance: ${:.2}\n", account.check_balance());

account.withdraw(300.0);
println!("New balance: ${:.2}\n", account.check_balance());

account.withdraw(2000.0);  // Will fail - insufficient funds
println!("Final balance: ${:.2}", account.check_balance());
```

### Comparison: Methods vs Associated Functions

| Feature | Methods | Associated Functions |
|---------|---------|----------------------|
| Has `self` parameter? | Yes | No |
| Called with | Dot notation: `instance.method()` | `::` syntax: `Struct::function()` |
| First parameter | `self`, `&self`, or `&mut self` | None - regular parameters |
| Use case | Operations on instance data | Constructors, utilities |
| Example | `circle.area()` | `Point::new(x, y)` |
| Can modify? | Only if `&mut self` | Depends on logic |

### Key Points About Structs

1. **Structs bundle related data** - Group fields that logically belong together
2. **Methods provide behavior** - Use `impl` blocks to define methods using `&self`, `&mut self`, or `self`
3. **Associated functions are constructors** - Use `::` syntax to call, commonly named `new()`
4. **Ownership works with structs** - Struct instances follow the same ownership rules as other values
5. **Structs implement patterns** - Enable object-oriented design patterns in Rust

Structs are the foundation of building larger, more complex Rust programs!


## Generics

Generics allow you to write code that can operate on **many different types** while still being
checked at compile time. The standard library uses them heavily – `Option<T>`, `Vec<T>`,
`Result<T, E>`, and many other types are generic.

You can add generics to:

- functions and methods
- structs and enums
- trait definitions and implementations (we'll get to traits later)

The syntax uses angle brackets with a type parameter name such as `<T>`, `<U>` or more
meaningful identifiers when appropriate. At compile time Rust performs *monomorphization*,
creating concrete implementations for each used type, so there is **no runtime cost**.

### Generic Functions

A generic function looks almost the same as a normal one, except the type parameter appears
between the function name and the parameter list. You can constrain the type with trait
bounds when you need specific behaviour.


```rust
// generic function that finds the largest element in a slice
// the type T must implement PartialOrd so we can compare values
// and Copy so we can return a value rather than a reference

fn largest<T: PartialOrd + Copy>(list: &[T]) -> T {
    let mut largest = list[0];
    for &item in list.iter() {
        if item > largest {
            largest = item;
        }
    }
    largest
}

let nums = vec![34, 50, 25, 100, 65];
println!("largest number is {}", largest(&nums));

let chars = vec!['y', 'm', 'a', 'q'];
println!("largest char is {}", largest(&chars));

// generic functions can also have multiple type parameters:
// fn something<T, U>(x: T, y: U) { ... }

// Generic Structs
// A struct with a type parameter is defined just like a function.
// Each instance can use a different concrete type.
/*
struct Point<T> {
    x: T,
    y: T,
}
/*
let int_point = Point { x: 5, y: 10 };
let float_point = Point { x: 1.0, y: 4.0 };
*/
// You can even mix types by using more than one parameter.
struct PointXY<T, U> {
    x: T,
    y: U,
}
/* 
let mixed = PointXY { x: 5, y: 4.0 };
*/
// Methods on generic structs also use generics:
impl<T> Point<T> {
    fn x(&self) -> &T {
        &self.x
    }
}

// the compiler will generate separate versions for i32 and f64 etc.

// Generic code is extremely powerful and keeps your APIs flexible!
*/
```

### Generics in Standard Library Types

Rust’s most commonly used collection and wrapper types are **generic**.
This is why `Option<T>`, `Vec<T>`, `Box<T>` and friends can hold *any* type:

- `Option<T>` can represent either `Some(value)` or `None`. The inner `T` is
  whatever type you need an integer, a string, even another generic type. OPtion is used as it can be NOne which is similar to nullptr or NULL in other languages.
- `Vec<T>` is a growable heap‑allocated array. The `T` parameter determines the
  element type stored inside.
- `Box<T>` is a smart pointer that allocates its contents on the heap. Using a
  generic parameter lets you box any sized type (especially useful for
  recursive structures).

These generics behave exactly like the ones we wrote above; the compiler
monomorphises concrete versions at compile time, so there’s no runtime cost.
Let’s try a few common examples.


```rust
// Option<T> example
let some_number: Option<i32> = Some(10);
let no_number: Option<i32> = None;

match some_number {
    Some(n) => println!("We got {}", n),
    None => println!("No value present"),
}

// Vec<T> example
let mut strings: Vec<String> = Vec::new();
strings.push(String::from("hello"));
strings.push(String::from("world"));
for s in strings.iter() {
    println!("vec contains: {}", s);
}

// Box<T> example - boxing a generic struct
struct Node<T> {
    value: T,
    next: Option<Box<Node<T>>>,
}

let leaf = Node { value: 5, next: None };
let tree = Node { value: 3, next: Some(Box::new(leaf)) };

println!("box holds value {}", tree.next.unwrap().value);

// You can also box primitive types or other collections
let boxed_str: Box<str> = Box::from("boxed slice");
println!("boxed string slice: {}", boxed_str);
```

## Traits

Traits define *shared behavior* that types can implement. They're similar to
interfaces in other languages but more powerful: they can provide default
method implementations, be used as bounds on generics, and support dynamic
dispatch via `dyn`.

Key points:

- Declare a trait with the `trait` keyword and specify required methods.
- Implement the trait for any type using `impl Trait for Type { ... }`.
- Traits can have **default method implementations** that types may override.
- Use trait bounds (`T: Trait`) to write generic functions that accept any type
  implementing the trait.
- `dyn Trait` allows for dynamic dispatch (useful when you need heterogeneous
  collections or runtime polymorphism).

Let's look at some examples.


```rust
// Define a trait
trait Summary {
    // required method
    fn summarize(&self) -> String;

    // default method - types implementing Summary automatically get this
    fn summarize_default(&self) -> String {
        String::from("(Read more...)")
    }
}

// Implement the trait for a struct
struct NewsArticle {
    headline: String,
    location: String,
    author: String,
    content: String,
}

impl Summary for NewsArticle {
    fn summarize(&self) -> String {
        format!("{} by {} ({})", self.headline, self.author, self.location)
    }
    // we could override summarize_default if we wanted
}

// Implement trait for another type
struct Tweet {
    username: String,
    content: String,
    reply: bool,
    retweet: bool,
}

impl Summary for Tweet {
    fn summarize(&self) -> String {
        format!("{}: {}", self.username, self.content)
    }
}

let tweet = Tweet {
    username: String::from("horse_ebooks"),
    content: String::from("Rust is awesome!"),
    reply: false,
    retweet: false,
};

let article = NewsArticle {
    headline: String::from("Rust 1.60 Released"),
    location: String::from("Internet"),
    author: String::from("Mozilla"),
    content: String::from("Lots of improvements..."),
};

println!("Tweet summary: {}", tweet.summarize());
println!("Article summary: {}", article.summarize());
println!("Default summary: {}", article.summarize_default());

// Using trait bounds with generics
fn notify<T: Summary>(item: &T) {
    println!("Breaking news! {}", item.summarize());
}

notify(&tweet);
notify(&article);

// Trait bound with where clause - more readable with multiple bounds
fn notify2<T, U>(t: &T, u: &U)
where
    T: Summary,
    U: Summary,
{
    println!("{} | {}", t.summarize(), u.summarize());
}

notify2(&tweet, &article);

// Using dyn Trait for dynamic dispatch
let feeds: Vec<Box<dyn Summary>> = vec![
    Box::new(tweet),
    Box::new(article),
];
for feed in feeds.iter() {
    println!("Feed: {}", feed.summarize());
}

```

> **Note on `Copy` and moves**
>
> Remember from earlier sections: `Copy` is itself a *trait* that certain
types implement (integers, bools, floats, chars, and tuples of `Copy` types).
> When a type implements `Copy`, the compiler duplicates the value on assignment
> instead of moving it. Types that do **not** implement `Copy` are moved by
default, transferring ownership and invalidating the original binding. These
> behaviours are not special cases – they follow the same trait system we’re
> discussing now. In fact, you can implement `Copy` (and `Clone`) for your own
types when it makes sense, giving them the familiar copy semantics.
